```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1=cv2.imread("5.png",1)
m2=cv2.inRange(m1, (240,240,240), (255,255,255))
m2=cv2.dilate(m2, np.ones((20,65)))
m3=np.full(m1.shape,255,np.uint8)
a,b=cv2.findContours(m2,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(m3,a,-1,(0,0,255),1)
print("輪廓數量：",len(a))

minYIndex=0
for i in range(1,len(a),1):
  x1, y1, w1, h1 =cv2.boundingRect(a[i-1])
  x2, y2, w2, h2 =cv2.boundingRect(a[i])
  if y2>y1:
    minYIndex=i

x, y, w, h =cv2.boundingRect(a[minYIndex])
cv2.rectangle(m1,(x,y),(x+w,y+h),(0,0,255),3)

cv2_imshow(m1)
# cv2_imshow(m2)
# cv2_imshow(m3)
```

```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from pyzbar import pyzbar
from PIL import ImageFont, ImageDraw, Image

m1=cv2.imread("10.png",1)

ret=pyzbar.decode(m1)
i=0
for d in ret:
  i+=1
  # print(d.type)
  x,y,w,h=d.rect
  cv2.rectangle(m1,(x,y),(x+w,y+h),(0,0,255),2)
  m1=Image.fromarray(m1)
  ImageDraw.Draw(m1).text((x+10,y+10), str(i), (0,255,0), ImageFont.truetype("mingliub.ttc", 50))
  m1=np.array(m1)
  print("編號：",i)
  try:
    print(d.data.decode("utf-8").encode("sjis").decode("utf-8"))
  except:
    print(d.data.decode("utf-8"))
  print("=============================")

cv2_imshow(m1)
```

```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from pyzbar import pyzbar
from PIL import ImageFont, ImageDraw, Image

m1=cv2.imread("face.jpg",1)

p1=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
ret=p1.detectMultiScale(m1,minNeighbors=3,minSize=(10,10))

for x,y,w,h in ret:
  cv2.rectangle(m1,(x,y),(x+w,y+h),(0,0,255),2)

cv2_imshow(m1)
```
