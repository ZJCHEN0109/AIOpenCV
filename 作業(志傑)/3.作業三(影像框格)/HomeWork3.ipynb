```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1=cv2.imread("3.png",1)
m2=cv2.inRange(m1, (250,250,250), (255,255,255))
m2=cv2.dilate(m2, np.ones((3,25)))
m3=np.full(m1.shape,255,np.uint8)
a,b=cv2.findContours(m2,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(m3,a,-1,(0,0,255),1)
print("輪廓數量：",len(a))

for i in range(0,len(a),1):
  x, y, w, h =cv2.boundingRect(a[i])
  cv2.rectangle(m1,(x,y),(x+w,y+h),(255,0,0),1)

cv2_imshow(m1)
# cv2_imshow(m2)
# cv2_imshow(m3)
```
