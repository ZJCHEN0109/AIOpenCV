```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont, ImageDraw, Image
import time

p1=cv2.VideoCapture("1.mp4")
while p1.isOpened()==True:
  ret, m1=p1.read()
  if ret==True:
    dp.clear_output(True)
    cv2.circle(m1, (20,20), 30, (0,255,0), -1)
    cv2_imshow(m1)
    time.sleep(0.3)
  else:
    break
```

```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont, ImageDraw, Image
import time

p1=cv2.VideoWriter("24.mp4", cv2.VideoWriter_fourcc(*'MP4V'), 30, (300,150))

for x in range(1,200,10):
  m4 = np.full((150,300,3), (255,255,255), np.uint8)
  cv2.rectangle(m4,(20+x,10), (120+x,120), (244,133,66), -1)
  dp.clear_output(True)
  cv2_imshow(m4)
  time.sleep(0.2)
  p1.write(m4)
for x in range(1,100,10):
  m5 = np.full((150,300,3), (255,255,255), np.uint8)
  cv2.rectangle(m5,(250-x,10), (120-x,120), (244,133,66), -1)
  dp.clear_output(True)
  cv2_imshow(m5)
  time.sleep(0.2)
  p1.write(m5)
p1.release()
```
